```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
data = numpy.loadtxt("tensor_bi-spectrum.dat")
```

```python
plt.cla()
plt.hold(True)
plt.xlabel(r'e-fold ${\rm N}$')
plt.ylabel(r'$P_{\rm T}$')
plt.title(r'$P_{\rm T}$ as a function of e-fold ${\rm N}$')
numerical, = plt.loglog(data[:,0], data[:,5], '*', label = 'numerical results')
plt.legend([numerical], ['numerical results'])
plt.savefig('power_spectrum_power_law.png')
plt.show()
```

```python
plt.cla()
plt.hold(True)
plt.xlabel(r'e-fold ${\rm N}$')
plt.ylabel(r'abs($\mathcal{G}$)')
plt.title(r'abs($\mathcal{G}$) as a function of e-fold ${\rm N}$')
numerical, = plt.loglog(data[:,0], numpy.sqrt(data[:,6]**2 + data[:,7]**2), '*', label = 'numerical results')
plt.legend([numerical], ['numerical results'])
plt.savefig('calG_vs_k_power_law.png')
plt.show()
```

```python
plt.cla()
plt.hold(True)
plt.xlabel(r'e-fold ${\rm N}$')
plt.ylabel(r'$k^6$abs($\mathcal{G}$)')
plt.title(r'$k^6$abs($\mathcal{G}$) as a function of e-fold ${\rm N}$')
numerical, = plt.semilogx(data[:,0], data[:,8], '*', label = 'numerical results')
plt.legend([numerical], ['numerical results'])
plt.savefig('k6calG_vs_k_power_law.png')
plt.show()
```

```python
plt.cla()
plt.hold(True)
plt.xlabel(r'e-fold ${\rm N}$')
plt.ylabel(r'$G$')
plt.title(r'$G$ as a function of e-fold ${\rm N}$')
numerical, = plt.loglog(data[:,0], data[:,9], '*', label = 'numerical results')
plt.legend([numerical], ['numerical results'])
plt.savefig('G_vs_k_power_law.png')
plt.show()
```

```python
plt.cla()
plt.hold(True)
plt.xlabel(r'e-fold ${\rm N}$')
plt.ylabel(r'$k^6G$')
plt.title(r'$k^6G$ as a function of e-fold ${\rm N}$')
numerical, = plt.loglog(data[:,0], data[:,10], '*', label = 'numerical results')
plt.legend([numerical], ['numerical results'])
plt.savefig('k6G_vs_k_power_law.png')
plt.show()
```

```python
plt.cla()
plt.hold(True)
plt.xlabel(r'e-fold ${\rm N}$')
plt.ylabel(r'$h_{\rm NL}$')
plt.title(r'$h_{\rm NL}$ as a function of e-fold ${\rm N}$')
numerical, = plt.semilogx(data[:,0], -data[:,11], '*', label = 'numerical results')
plt.legend([numerical], ['numerical results'])
plt.savefig('h_NL_vs_k_power_law.png')
plt.show()
```
